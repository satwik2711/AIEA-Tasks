```python
from llama_index.readers.web import SimpleWebPageReader


documents = SimpleWebPageReader(html_to_text=True).load_data(
    ["https://www.formula1.com/en/results.html/2024/drivers.html"]
)
```

```python
documents
```

```python
from llama_index.core import VectorStoreIndex,  Settings
from llama_index.embeddings.ollama import OllamaEmbedding
from llama_index.llms.ollama import Ollama
```

```python
ollama_embedding = OllamaEmbedding(
    model_name="llama3",
    base_url="http://localhost:11434",
    ollama_additional_kwargs={"mirostat": 0},
)
Settings.embed_model = ollama_embedding
```

```python
Settings.llm = Ollama(model="llama3")
```

```python
index = VectorStoreIndex.from_documents(
    documents,
)
```

```python
query_engine = index.as_query_engine()
response = query_engine.query("who is the best formula 1 driver in 2024?")
print(response)
```

```python
query_engine = index.as_query_engine()
response = query_engine.query("who is the better ferrari driver? Also tell if Lewis Hamilton joins Ferrari next year instead of Carlos Sainz, who will be the better Ferrari driver in 2025?")
print(response)
```
